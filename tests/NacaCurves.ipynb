```python
x = np.linspace(0, 1, 11)
thickness = NacaCurves._createThicknessDistribution(4, x)
print(thickness)
```

```python
fig = figure(1);
ax  = fig.add_subplot(111, aspect='equal')
plot(thickness[:,0], thickness[:,1], 'bo-', label="thickness")
show()
```

```python
naca = NacaCurves.create4DigitNacaAerofoil(0, 0, 12, x);
upper = naca['upper'];
lower = naca['lower'];
camber = naca['camber'];
anticlockwise = naca['anticlockwise']
subplot(1, 2, 1)
plot(upper[:,0], upper[:,1])
plot(lower[:, 0], lower[:, 1])
plot(camber[:, 0], camber[:, 1])
subplot(1, 2, 2)
plot(anticlockwise[:,0], anticlockwise[:,1])
show()
```
